```python
df = pd.concat([df, pd.get_dummies(df['Medal'])], axis=1)
```

```python
#We have to remove the duplicates on the basis of the [NOC	Games	Year	Season	City	Sport]
medal_tally = df.drop_duplicates(subset=['NOC','Games','Year','Season','City','Sport','Event','Medal'])
```

```python
medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending = False).reset_index()
```

```python
#Now our anlayzed data is closed to reality
medal_tally = medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending=False).reset_index()
```

```python
#adding the gold and silver 
medal_tally['Total'] = medal_tally['Gold'] + medal_tally['Silver'] + medal_tally['Bronze']
```
