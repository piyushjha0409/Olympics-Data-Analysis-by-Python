```python
import numpy as np
import pandas as pd
```

```python
players_df = pd.read_excel(r"C:\Users\Default\Desktop\Data Analytics\project 1(Data Anlysis)\Athletes.xlsx")
```

```python
#Name of all the players
list(players_df['Name'])
```

```python
#converting the data into contigency tables with frequency
contigency = pd.crosstab(players_df['Name'], players_df['NOC'], normalize='all')
contigency
```

```python
#List of all the disciplines 
list(players_df['Discipline'])
```

```python
players_df.groupby(['Discipline']).count()
```

```python
players_df.groupby(['NOC']).count()
```

```python
#frequency table
freq_table = players_df.groupby(['Discipline']).size().reset_index(name='Count').rename(columns={'Discipline':'Game'})
freq_table
```
